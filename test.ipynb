```python
import os
import json
import pandas as pd
```

```python
# Aggregated Transaction

path1 = '/Users/prasanth/Desktop/ds_projects/Phonepe/pulse/data/aggregated/transaction/country/india/state/'
agg_trans_list = os.listdir(path1)

columns1={"States":[],"Years":[],"Quarter":[],"Transaction_type":[],"Transaction_count":[],"Transaction_amount":[]}

for state in agg_trans_list:
    cur_state = os.path.join(path1, state)
    if not os.path.isdir(cur_state):  # Check if it's a directory
        continue
    agg_year_list = os.listdir(cur_state)
    for year in agg_year_list:
        cur_year = os.path.join(cur_state, year)
        if not os.path.isdir(cur_year):  # Check if it's a directory
            continue
        agg_file_list = [file for file in os.listdir(cur_year) if not file.startswith('.')]  # Skip files starting with '.'
        
        for file in agg_file_list:
            cur_file = os.path.join(cur_year, file)
            try:
                with open(cur_file, "r") as data_file:
                    A = json.load(data_file)
                    for i in A["data"]["transactionData"]:
                        name = i["name"]
                        count = i["paymentInstruments"][0]["count"]
                        amount = i["paymentInstruments"][0]["amount"]
                        columns1["Transaction_type"].append(name)
                        columns1["Transaction_count"].append(count)
                        columns1["Transaction_amount"].append(amount)
                        columns1["States"].append(state)
                        columns1["Years"].append(year)
                        columns1["Quarter"].append(int(file.strip(".json")))
            except FileNotFoundError:
                print(f"File not found: {cur_file}")
            except Exception as e:
                print(f"Error processing file {cur_file}: {e}")
```

```python
agg_transaction = pd.DataFrame(columns1)
```

```python
# Aggregated user

path2 = "/Users/prasanth/Desktop/ds_projects/Phonepe/pulse/data/aggregated/user/country/india/state"

agg_user_list = os.listdir(path2)

columns2 = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "Brands": [],
    "Transaction_count": [],
    "Percentage": []
}

for state in agg_user_list:
    cur_state = os.path.join(path2, state) + "/"  # Add a slash to ensure correct path formatting
    if not os.path.isdir(cur_state):  # Check if it's a directory
        continue
    agg_year_list = os.listdir(cur_state)

    for year in agg_year_list:
        cur_year = os.path.join(cur_state, year) + "/"  # Add a slash to ensure correct path formatting
        if not os.path.isdir(cur_year):  # Check if it's a directory
            continue
        # Skip .DS_Store files
        agg_file_list = [file for file in os.listdir(cur_year) if not file.startswith('.')] 

        for file in agg_file_list:
            cur_file = os.path.join(cur_year, file)
            try:
                with open(cur_file, "r") as data_file:
                    B = json.load(data_file)
                    
                    try:

                        for i in B["data"]["usersByDevice"]:
                            brand = i["brand"]
                            count = i["count"]
                            percentage = i["percentage"]
                            columns2["Brands"].append(brand)
                            columns2["Transaction_count"].append(count)
                            columns2["Percentage"].append(percentage)
                            columns2["States"].append(state)
                            columns2["Years"].append(year)
                            columns2["Quarter"].append(int(file.strip(".json")))

                    except:
                        pass
                        
            except FileNotFoundError:
                print(f"File not found: {cur_file}")
            except Exception as e:
                print(f"Error processing file {cur_file}: {e}")
```

```python
# Map User

path4="/Users/prasanth/Desktop/ds_projects/Phonepe/pulse/data/map/user/hover/country/india/state"

map_user_list = os.listdir(path4)

columns4 = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "Districts": [],
    "RegisteredUsers": [],
    "AppOpens": []
}

for state in map_user_list:
    cur_state = os.path.join(path4, state) + "/"  # Add a slash to ensure correct path formatting
    if not os.path.isdir(cur_state):  # Check if it's a directory
        continue
    agg_year_list = os.listdir(cur_state)

    for year in agg_year_list:
        cur_year = os.path.join(cur_state, year) + "/"  # Add a slash to ensure correct path formatting
        if not os.path.isdir(cur_year):  # Check if it's a directory
            continue
        # Skip .DS_Store files
        agg_file_list = [file for file in os.listdir(cur_year) if not file.startswith('.')] 

        for file in agg_file_list:
            cur_file = os.path.join(cur_year, file)
            try:
                with open(cur_file, "r") as data_file:
                    D = json.load(data_file)
                    
                    for i in D["data"]["hoverData"].items():
                        district = i[0]
                        registeredUsers = i[1]["registeredUsers"]
                        appOpens = i[1]["appOpens"]
                        columns4["Districts"].append(district)
                        columns4["RegisteredUsers"].append(registeredUsers)
                        columns4["AppOpens"].append(appOpens)
                        columns4["States"].append(state)
                        columns4["Years"].append(year)
                        columns4["Quarter"].append(int(file.strip(".json")))
    
            except:
                pass
```
